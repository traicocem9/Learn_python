### SoftRelu 

$$f(x) = log(1+e^x)$$

```python
x = plt.linspace(-10,10,100)
a = 1
b = 1
y = a*x+b

y_2 = np.log(1+np.exp(y))
plt.plot(x,y_2, label = 'soft relu')
```

## Leaky Relu

$$ f(x) =\begin{cases} 0.01x, \mbox{for } x< 0\\x, \mbox{for } x >= 0 \end{cases}  $$

```python
x = plt.linspace(-10,10,100)
a = 1
b = -9
y = a*x+b
def leakyrelu():
    i = len(y)
    for i in range (i):
        if y[i] < 0:
            y[i] = 0.01*y[i]
        else:
            y[i] = y[i]
    return y
y = leakyrelu()
plt.plot(x,y,'r')
```

## Binary step

$$ f(x) =\begin{cases} 0, \mbox{for } x< 0\\1, \mbox{for } x >= 0 \end{cases}  $$

```python
x = plt.linspace(-10,10,100)
a = 2
b = 2
y = a*x+b
def binary_step():
    i = len(y)
    for i in range (i):
        if y[i] < 0:
            y[i] = 0
        else:
            y[i] = 1
    return y
y = binary_step()
plt.plot(x,y,'r')
```

## Inverse square root linear unit


$$   f(,x)=\begin{cases} \frac{x}{\sqrt{1+ax^2}}, \mbox{for } x< 0 \\ x, \mbox{for } x >= 0 \end{cases}  $$

```python
x = plt.linspace(-10,10,100)
a = 5
b = 1
y = a*x+b
alpha = 0.1
def Inverse_square_root_linear_unit():
    i = len(y)
    for i in range (i):
        if y[i] < 0:
            y[i] = y[i] / (np.sqrt(1+(alpha*y[i])**2))
        else:
            y[i] = y[i]
    return y
y = Inverse_square_root_linear_unit()
plt.plot(x,y,'r')
```

## ELU 

$$   f(,x)=\begin{cases} (e^x+1) , \mbox{for } x \leq 0 \\ x, \mbox{for } x > 0 \end{cases}  $$

```python
x = plt.linspace(-10,10,100)
a = 0.4
b = -1
y = a*x+b
alpha = 1
def ELU():
    i = len(y)
    for i in range (i):
        if y[i] <= 0:
            y[i] = (alpha*(np.exp(y[i])-1))
        else:
            y[i] = y[i]
    return y
y = ELU()
plt.plot(x,y,'r')
```
