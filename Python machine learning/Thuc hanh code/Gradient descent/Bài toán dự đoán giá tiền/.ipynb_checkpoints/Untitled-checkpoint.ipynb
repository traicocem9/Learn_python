```python
x = df.values[:,2]
y = df.values[:,4]
x_ = np.mean(x)
y_ = np.mean(y)
def m(x,y,x_,y_):

    return sum((x - x_)*(y - y_)) / sum(np.square(x-x_))
m = m(x,y,x_,y_)        
def c(x,y,x_,y_,m):

    return y_-m*x_
c = c(x,y,x_,y_,m)
print (c)
a = (m,c)

plt.scatter(x,y)

plt.plot(x,np.polyval(a,x))
```
