```python
# Gradient descent
def gd (X ,y ,w_ht ,b_ht ,learning_rate):
    n = len(X.T)
    w_dh = [0,0]
    b_dh = 0
    lost = 0
    for p in range (1000):
        for i in range (n):
            temp = X[:,i]
            tich = np.dot(w_ht, temp)
            w_dh += -2*temp.T*(y[i] - (tich + b_ht))
            b_dh += -2* (y[i] - (tich + b_ht))
            
            lost += (y[i] - (tich + b_ht))**2
        w_dh = w_dh / n
        b_dh = b_dh / n

        w_ht -= w_dh * learning_rate
        b_ht -= b_dh * learning_rate
        
        print('lost:', lost )
    
    return w_ht, b_ht, lost

def cost_function (X, y, w_ht, b_ht):
    n = len(x)
    error = 0
    
    for i in range(n):
        return sum()
    pass
```

```python
def simple_linear_regression(x_train, y_train):
    SLL = 100
    learning_rate = 0.01
    
    n = len(x_train)
    
    My = np.max(y_train)**2+1
    Mx = np.max(x_train)**2+1
    
    y = y_train/My
    x = x_train/Mx
    a_0 = 0
    
    a_1 = np.array([1,1])
    
    cost_his = []
    for i in range (SLL):
        
        
        
        y_pred = a_0 + np.dot(a_1,x)
        error = y - y_pred
        
        D_a_1_1 = x * (y - y_pred)
        D_a_0_1 = y-y_pred
        
        D_a_1 = (-2/n) * D_a_1_1.sum(axis = 1)
        D_a_0 = (-2/n) * D_a_0_1.sum(axis = 0)  
        
        a_1 = a_1 - learning_rate * D_a_1
        a_0 = a_0 - learning_rate * D_a_0  
        
        cost = (1/n)*sum((y - y_pred)**2)
        print('cost {}; SLL {}'.format(cost,i))
        #print (np.shape(D_a_1_1))
        #print (a_1)
        cost_his.append(cost)
    return a_1*My/Mx, a_0*My, SLL, cost_his
    
    
# test 1
#x_train = np.array([1, 2, 3, 4, 5])
#y_train = np.array([5, 7, 9, 11, 13])
# test 2
#x_train = np.array([147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183])
#y_train = np.array([ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68])
#train 3
x_train = np.array(X)
y = df.values[:,4]
y_train = np.array(y)

a_1, a_0, SLL, cost = simple_linear_regression(x_train, y_train)

a = np.array([ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68])
#plt.scatter(x_train.T[1],y_train)
plt.plot(cost,color = 'r')
#plt.legend(['y (true)', 'y (predicted)'], loc='upper left')
plt.show()
```
