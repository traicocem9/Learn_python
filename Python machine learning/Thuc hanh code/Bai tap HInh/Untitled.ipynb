### matplotlib


```python
from numpy import random
import numpy as np
import random
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
#import Hinh
img_dir = 'img.png'

image = img.imread(img_dir)

plt.subplot(231)
plt.imshow(image)
plt.show()

print(image)
print(np.shape(image))
```

```python
# Dao hinh 90 cung chieu kim dong ho
X = np.rot90(image,-1)
plt.imshow(X)

```

```python
# Nguoc chieu kim dong ho
Y = np.rot90(image, 1)
plt.imshow(Y)
```
