## dữ liệu ở trên sẽ dc chuyển sang dạng số để dễ thực hiện tính toán
$$nhẹ = 1

#### tạo 1 ma trận 2 chiều (X) bên trong mỗi 1 ma trận chứa dữ liệu tương đương với số thứ tự 1-10
## Lưu ý: không nhập dữ liệu 9 và 10 vì đó là dữ liệu mà chúng ta sẽ dự đoán


```python
X = [[1,3,3,7],[5,2,4,6],[5,2,4,6],[1,2,4,6],[5,4,4,3],[1,4,4,7],[3,2,3,7],[3,3,3,6]]
```

#### và (Y) là ma trận chứa kết quả (bệnh tim) của từng ma trận ở trên 


```python
'số phòng' :[1,3,1,3,1,2,2,3,1,1], 'diện tích nhà':[21,43,22,40,19,29,34,41,20,21] ,'Sân vườn':[0,1,1,0,0,0,0,1,0,0]
```

## Tạo 1 cây dự đoán xử lý dữ liệu từ (X) và (Y)
Chúng ta sẽ lấy dữ liệu từ (X) và (Y) đưa vào lệnh. hệ thống sẽ tự động tính toán 
### Tương tự Root node là (Y) , Interior node là (X), Leaf node là dữ liệu bên trong (X)



```python
Test = tree.fit(X,Y)
```

## Khi hệ thống đã xử lí xong dữ liệu ta bắt đầu dự đoán kết quả đầu ra (bệnh tim)  . Bắt đầu nhập dữ liệu 9-10 để hệ thống dự đoán kết quả là bệnh tim 

```python
predict = Test.predict([[1,4,3,6],[1,4,3,7]])
predict
```
