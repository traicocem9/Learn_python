```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
# doc data trong file csv 
df = pd.read_csv('Advertising.csv')
# lay cot 2 va 4 trong data
x = df.values[:,2]
y = df.values[:,4]
```

```python
# tao ra cac diem du lieu 
plt.scatter(x,y, marker ='o')
```

```python
#ham du doan gia tri 
def predict(new_radio, weight, bias):
    return weight*new_radio + bias

#ham` sai so
def cost_function(x,y, weight, bias):
    n = len(x)
    sum_error = 0
    for i in range(n):
        y_dudoan = (weight*x[i]+bias)

        sum_error += (y[i] - y_dudoan)**2
    
    return sum_error/n
    return sum_error
```

## Cng thc o hm 


```python
def update_weight(x, y, weight, bias, learning_rate):
    n = len(x)
    weight_temp = 0
    bias_temp = 0
    for i in range (n):
        weight_temp = (2*x[i]*(y[i]-(weight*x[i]+bias)))/n
        bias_temp = (2*(y[i] - (weight*x[i]+bias)))/n
    weight -= learning_rate * weight_temp
    bias -= learning_rate * bias_temp
    return weight, bias
```

```python

def train(x,y,weight,bias,learning_rate, iter):
    cost_his = []
    for i in range (iter):
        weight, bias = update_weight(x, y, weight, bias, learning_rate)
        cost = cost_function(x,y, weight, bias)
        cost_his.append(cost)
        
    return weight, bias, cost_his

weight, bias, cost = train(x, y, 0.1, 0.14, 0.001, 20)

print('ket qua:')
print(weight)
print (bias)
print(cost)
predict(x, weight, bias)
```
